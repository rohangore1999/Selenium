```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import pickle
import numpy as np
import json

import re
from collections import OrderedDict
from time import strptime

import sys
```

```python
file = open("input.txt", "r")
c = file.read()
#skill
skill = c.split("\n")[0].split(":")[1]
#location 
location = c.split("\n")[1].split(":")[1]

#filter
#remote option
remote_option = c.split("\n")[2].split(":")[1]

#posted date
posted_date = c.split("\n")[3].split(":")[1]

#emp type
emp_type =  c.split("\n")[4].split(":")[1]

#distance
dist = c.split("\n")[5].split(":")[1]

#empr type
empr_type = c.split("\n")[6].split(":")[1]


#result_id
result_id = 0

# list of dictionary
main = []

# Initializing data Frame
df = pd.DataFrame()

driver = webdriver.Chrome(executable_path=r"chromedriver.exe")
driver.get("https://www.dice.com/")

#enter skill
driver.find_element_by_xpath('//*[@id="searchInput-div"]/form/div/div[1]/div/dhi-new-typeahead-input/div/input').clear()
sleep(2)
driver.find_element_by_xpath('//*[@id="searchInput-div"]/form/div/div[1]/div/dhi-new-typeahead-input/div/input').send_keys(skill)

#enter location
driver.find_element_by_xpath('//*[@id="google-location-search"]').clear()
sleep(2)
driver.find_element_by_xpath('//*[@id="google-location-search"]').send_keys(location)

#submit
driver.find_element_by_xpath('//*[@id="submitSearch-button"]').click()

sleep(10)

#check for filters if any then apply it
print("FILTERS: ")
try:
    # REMOTE OPTION
    if remote_option.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[1]/li[1]').text.replace(" ","").lower():
        print("remote only")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[1]/li[1]').click()

    elif remote_option.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[1]/li[2]').text.replace(" ","").lower():
        print("Exclude Remote")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[1]/li[2]').click()

    elif  remote_option.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[2]').text.text.replace(" ","").lower():
        print("Work From Home Available")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[1]/div[2]/div/js-remote-options-filter/div/ul[2]').click()
except:
    pass

sleep(10)

try:
    #POSTED DATES
    if posted_date.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[1]').text.replace(" ","").lower():
        print("any date")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[1]').click()

    elif posted_date.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[2]').text.replace(" ","").lower():
        print("today")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[2]').click()

    elif  posted_date.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[3]').text.replace(" ","").lower():
        print("last 3 days")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[3]').click()

    elif posted_date.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[4]').text.replace(" ","").lower():
        print("last 7 days")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[2]/div[2]/div/js-single-select-filter/div/div/button[4]').click()

except:
    pass

sleep(10)

try:    
    #EMPLOYMENT TYPE
    if emp_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[1]').text.replace(" ","").lower():
        print("Fill time")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[1]').click()

    elif emp_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[2]').text.replace(" ","").lower():
        print("part time")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[2]').click()

    elif  emp_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[3]').text.replace(" ","").lower():
        print("contract")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[3]').click()

    elif emp_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[4]').text.replace(" ","").lower():
        print("third party")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[3]/div[2]/div/js-multi-select-filter/ul/li[4]').click()

except:
    pass

sleep(10)

try:
    #DISTANCE
    #clear
    driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[4]/div[2]/div/dhi-radius-filter-widget/dhi-radius-filter/form/div[1]/div/input').clear()

    #enter new distance
    driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[4]/div[2]/div/dhi-radius-filter-widget/dhi-radius-filter/form/div[1]/div/input').send_keys(dist.split(" ")[1])
except:
    pass

sleep(10)

try:
    #EMPLOYER TYPE
    if empr_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[5]/div[2]/div/js-multi-select-filter/ul/li[1]').text.replace(" ","").lower():
        print("recruiter")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[5]/div[2]/div/js-multi-select-filter/ul/li[1]').click()

    elif empr_type.replace(" ","").lower() in driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[5]/div[2]/div/js-multi-select-filter/ul/li[2]').text.replace(" ","").lower():
        print("direct hire")
        driver.find_element_by_xpath('//*[@id="facets"]/dhi-accordion[5]/div[2]/div/js-multi-select-filter/ul/li[2]').click()
except:
    pass

print(" ")

total = driver.find_element_by_xpath('//*[@id="totalJobCount"]').text
total = int(int(total)/20)
print("total: ", total)

#pages upto 9
for p in range(total+1):
    try:
        #for title
        for t in range(0,19):
            result_id += 1
            try:
                #title(0-19)
                try:
                    title = driver.find_element_by_xpath('//*[@id="searchDisplay-div"]/div[2]/dhi-search-cards-widget/div').find_elements_by_class_name('search-card')[t].find_element_by_class_name('card-title-link').text
                except:
                    title = 'None'

                #link
                try:
                    link = driver.find_element_by_xpath('//*[@id="searchDisplay-div"]/div[2]/dhi-search-cards-widget/div').find_elements_by_class_name('search-card')[t].find_element_by_class_name('card-title-link').get_attribute('href')
                    driver.get(link)
                except:
                    pass

                #employer
                try:
                    employer = driver.find_element_by_class_name('employer').text
                except:
                    employer = "None"

                #location
                try:
                    location = driver.find_element_by_class_name('location').text
                except:
                    location = "None"
                
                try:
                    #dicer id
                    dicer_id = driver.find_element_by_xpath('//*[@id="bd"]/div/div[2]/div[1]/div[3]/div').text.split(':')[1]
                except:
                    dicer_id = 'None'
                
                try:
                    #dicer id
                    dicer_id = driver.find_element_by_xpath('//*[@id="bd"]/div/div[2]/div[1]/div[3]/div').text.split(':')[1]
                except:
                    dicer_id = 'None'
                
                try:
                    #position id
                    position_id = driver.find_element_by_xpath('//*[@id="bd"]/div/div[2]/div[1]/div[4]/div').text.split(':')[1]
                except:
                    position_id = 'None'
                    
                #when_poster
                try:
                    when_poster = driver.find_element_by_class_name('posted').text
                except:
                    when_poster = "None"

                #required_skill
                try:
                    required_skill = driver.find_elements_by_class_name('job-info')[0].text
                except: 
                    required_skill = "None"

                #employee_type
                try:
                    employee_type = driver.find_elements_by_class_name('job-info')[1].text
                except:
                    employee_type = "None"

                #money
                try:
                    money = driver.find_elements_by_class_name('job-info')[2].text
                except:
                    money = "None"

                #travel
                try:
                    travel = driver.find_elements_by_class_name('job-info')[3].text
                except:
                    travel = "None"

                #job_description
                try:
                    job_description = driver.find_element_by_id('jobdescSec').text.replace("\n"," ")
                except:
                    job_description = "None"

                driver.back()
                
                sleep(10)
                
                print("making dict")
                #making dictonary
                data ={
                 'result_id': result_id,
                 'Job Ttle':title,
                 'employer':employer,
                 'location':location,
                 'dicer_id':dicer_id,
                'position_id': position_id,
                 'when_poster':when_poster,
                 'required_skill': required_skill,
                 'employee_type': employee_type,
                 'money' : money,
                 'travel' : travel,
                 'job_description':job_description
                 }
    #                     print(data)

                #append it to main data
                main.append(data)
                print("Making JSON")
                #Open and save in json file
                with open("demo.json", 'w', encoding="utf-8") as f:
                          json.dump(main, f,indent=4, ensure_ascii=False)
                
                #creating dataframe
                df = df.append({
                 'result_id': result_id,
                 'title':title,
                 'employer':employer,
                 'location':location,
                 'dicer_id':dicer_id,
                 'position_id': position_id,
                 'when_poster':when_poster,
                 'required_skill': required_skill,
                 'employment_type': employee_type,
                 'money' : money,
                 'travel' : travel,
                 'job_description':job_description
                 }, ignore_index=True)

                print("Making csv")
                df.to_csv('input_dataframe.csv')
            except Exception as e:
                print(e)
                pass
    except:
        pass
    
    #click to next page
    try:
        driver.find_element_by_xpath('//*[@id="pagination_2"]/pagination/ul/li[7]/a').click()
        sleep(10)
    except Exception as e:
#         print(e)
        break
    
```
