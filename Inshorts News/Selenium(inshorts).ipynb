```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from fuzzywuzzy import fuzz
import pickle
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import gensim
import json
```

```python
headlines=[]
body=[]
img_src=[]
time=[]
date=[]
headlines1=[]
dictmain={}
driver = webdriver.Chrome(executable_path=r"chromedriver.exe")
driver.get("https://inshorts.com/en/read")
for j in range(1,100):
    
    #make dict
    dic={}
    
    try:
        #headlines
        res=driver.find_element_by_xpath('/html/body/div[4]/div/div[3]/div['+str(j)+']/div/div[2]/a/span').text
        headlines.append(res)
        
        #imgsrc
        img=driver.find_element_by_xpath('/html/body/div[4]/div/div[3]/div['+str(j)+']/div/div[1]').get_attribute("style")
        img_src.append(img[23:])
        
        #body
        b=driver.find_element_by_xpath('/html/body/div[4]/div/div[3]/div['+str(j)+']/div/div[3]/div[1]').text
        body.append(b)
        
        #time
        t=driver.find_element_by_xpath('/html/body/div[4]/div/div[3]/div['+str(j)+']/div/div[2]/div/span[3]').text
        time.append(t)
        
        #date
        d=driver.find_element_by_xpath('/html/body/div[4]/div/div[3]/div['+str(j)+']/div/div[2]/div/span[2]').text
        date.append(d)
        
        
        dic['headline']=res
        dic['img_src']=img
        dic['body']=b
        dic['time']=t
        dic['date']=d
        
        
        
        dictmain[j]=dic
        results=json.dumps(dictmain)

    except:
        pass

```
